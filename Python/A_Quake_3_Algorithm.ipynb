```python
import numpy as np 
import matplotlib.pyplot as plt 

dt = 0.01
t = np.arange(0,1,dt)
f = np.sin(2*np.pi*50*t)+np.sin(2*np.pi*120*t)
f_clean = f
f = f + 2.5*np.random.randn(len(t))

plt.plot(t,f,color = 'r',LineWidth = 1.5,label = 'Noisy')
plt.plot(t,f_clean,color = 'w',LineWidth = 2,label = "Clean")
plt.xlim(t[0],t[-1])
plt.legend()
```
